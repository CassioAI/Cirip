```python
import requests
import pandas as pd
from PIL import Image
import keras.preprocessing.image as k_image
import numpy as np
import random
import os
import skimage.color as skcolor
from sklearn.preprocessing import StandardScaler
```

```python
def is_grey_scale(im):
    w,h = im.size
    rand_pixel_array = np.zeros((10,2))
    for i in range(10):
        rand_pixel_array[i,:] = [random.randint(0,w-1), random.randint(0,h-1)]
    for i in range(10):
        r,g,b = im.getpixel((rand_pixel_array[i,0], rand_pixel_array[i,1]))
        if r != g != b: return False
    return True
```

```python
def has_proper_dim(im):
    im_array = np.array(im)
    shape = im_array.shape
    if((len(shape) != 3) or (shape[2] != 3)):
        return False
    else:
        return True
    
```

```python
def download_image(url, filepath, filename):
    full_path = filepath + filename
    try:
        r = requests.get(url, allow_redirects=False)
    except requests.exceptions.RequestException as e:
        print(e)
        return False
    if(r.status_code == 200):
        with open(full_path, 'wb') as f:
            f.write(r.content)
        im = Image.open(full_path);
        if(has_proper_dim(im) == True):
            if(is_grey_scale(im) == False):
                return True
            else:
                print(' greyscale error')
                os.remove(full_path)
                return False
        else:
            print(' dimension error')
            os.remove(full_path)
            return False
    else:
        return False
```

```python
def crop_resize_Image(img_path, target_size):
    im = Image.open(img_path)
    width,height = im.size
    if(width > height):
        top     = 0
        left    = int((width - height)/2)
        bottom  = height
        right    = width - int((width - height)/2)
    else:
        top     = int((height - width)/2)
        left    = 0
        bottom  = height-int((height - width)/2)
        right    = width
    im = im.crop((left,top,right,bottom)).resize(target_size)
    im.save(img_path)
```

```python
def standardize(array):
    ave = np.average(array)
    std = np.std(array)
    array = (array-ave)/std
    return array,ave,std
```

```python
def download_img_dataset(csv_filepath, target_filepath, num_imgs, offset, separator, url_col_idx ,num_rows_skip):
    df = pd.read_csv(csv_filepath, sep = separator, skiprows = num_rows_skip)
    num_downl_imgs = 0
    idx = 0
    while num_downl_imgs < num_imgs:
        if(download_image(df.iloc[idx,url_col_idx], target_filepath, df.iloc[idx,url_col_idx] + '.jpg') == True):
            crop_resize_Image(target_filepath + df.iloc[idx,url_col_idx] + '.jpg',(256,256))
            num_downl_imgs += 1
            idx += 1
        else:
            idx += 1    
```
